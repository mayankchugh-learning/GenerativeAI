```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0
    )
    return response.choices[0].message.content
```

```python
res = get_completion("who won the first cricket worldcup?")
print(res)
```

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {
          "role": "system", "content": "you are my assitant",
          "role": "user", "content": "who won the first cricket worldcup?"
        }
      ]
    ,
    max_tokens=150,
)
```
